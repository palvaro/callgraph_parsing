```python
from cal2_parser import CalParser
from graphviz import Digraph
```

```python
p = CalParser("viewitem.json")
roots = p.process()
```

```python
def draw_graph(tree, name, rule=[]):
    dot = Digraph(name)
    tree.todot(dot, rule)
    return dot
```
